```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
data_set = pd.read_csv("E:\SAIF\prwatech\ml\Datasets\Test_Dataset\SUV_predict.csv")
```

```python
X = data_set.iloc[:,[2,3]].values
```

```python
X
```

```python
y = data_set.iloc[:,[4]].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=200)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_train
```

```python
X_test = sc.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=200)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
z=classifier.predict([[24,100000]])
```

```python
z
```

```python
y_predict
```

```python
#confusion matrix or accuracy_score

from sklearn.metrics import accuracy_score
```

```python
cm = accuracy_score(y_test,y_predict)*100
```

```python
cm
```
